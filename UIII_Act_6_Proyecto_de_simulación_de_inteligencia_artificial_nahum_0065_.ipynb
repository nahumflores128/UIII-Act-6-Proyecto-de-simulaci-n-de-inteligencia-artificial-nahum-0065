###Descripción de las líneas y conceptos utilizados:
### Ciclo while interactivo: El programa utiliza una variable de control (continuar) para repetir el bloque de código mientras la condición sea verdadera, permitiendo múltiples evaluaciones sin reiniciar el script
.
### Conversión de tipos (int, float): Las entradas del usuario son originalmente cadenas de texto; se convierten a números para poder realizar comparaciones de magnitud y operaciones matemáticas
.
### Operadores Aritméticos: Se utiliza la resta (-) para determinar el flujo de caja libre y la multiplicación (*) para calcular el porcentaje del monto solicitado que el cliente puede cubrir
.
### 0peradores Relacionales: Se usan comparaciones como menor que (<), mayor que (>) e igualdad (==) para clasificar al solicitante en diferentes niveles de riesgo
.
### Operadores Lógicos: El uso de and permite que el sistema exija el cumplimiento de dos condiciones simultáneas (ej. tener dinero disponible Y no tener atrasos) para otorgar la mejor puntuación
.
### Estructura if-elif-else: Funciona como el motor de "inteligencia" al encadenar múltiples decisiones posibles, evaluando desde el escenario más optimista hasta el rechazo por defecto
.

```python
# --- SIMULADOR DE INTELIGENCIA LÓGICA: BURÓ DE CRÉDITO ---

# Variable de control para mantener el programa activo de forma interactiva
continuar = "si"

while continuar == "si":
    print("\n--- Sistema de Evaluación de Crédito Proactivo ---")

    # 1. Entrada de datos interactiva y conversión de tipos
    # Se solicita información y se convierte a entero o flotante para operar con ellos
    nombre = input("Ingrese el nombre del solicitante: ")
    edad = int(input("Ingrese la edad: "))
    ingresos = float(input("Ingrese ingresos mensuales (USD): "))
    gastos = float(input("Ingrese gastos mensuales fijos (USD): "))
    tiene_atrasos = input("¿Tiene historial de atrasos en pagos? (si/no): ")
    monto_solicitado = float(input("Monto del crédito solicitado: "))

    # 2. Procesamiento de IA mediante lógica condicional
    # Primero se verifica un requisito indispensable: la mayoría de edad
    if edad < 18:
        print(f"\n[RESULTADO]: Solicitud rechazada.")
        print("Razón: El solicitante debe ser mayor de edad.")

    else:
        # Cálculo aritmético de la capacidad de pago mensual
        # Se resta el gasto de los ingresos para obtener el saldo disponible
        capacidad_pago = ingresos - gastos

        # Evaluación de riesgo utilizando operadores relacionales y lógicos
        # Se considera 'Excelente' si tiene buena capacidad y no tiene atrasos
        if capacidad_pago > (monto_solicitado * 0.2) and tiene_atrasos == "no":
            puntuacion = 900
            nivel_riesgo = "Mínimo"
            decision = "APROBADO con tasa preferencial"

        # Se considera 'Bueno' si la capacidad es moderada y sin atrasos
        elif capacidad_pago > (monto_solicitado * 0.1) and tiene_atrasos == "no":
            puntuacion = 700
            nivel_riesgo = "Bajo"
            decision = "APROBADO con tasa estándar"

        # Se considera 'Regular' si tiene capacidad pero cuenta con atrasos
        elif capacidad_pago > (monto_solicitado * 0.1) and tiene_atrasos == "si":
            puntuacion = 500
            nivel_riesgo = "Medio"
            decision = "APROBADO sujeto a garantía prendaria"

        # Caso por defecto: poca capacidad o historial muy negativo
        else:
            puntuacion = 300
            nivel_riesgo = "Alto"
            decision = "RECHAZADO por insuficiencia financiera"

        # 3. Visualización de resultados finales
        print(f"\n--- REPORTE PARA: {nombre} ---")
        print(f"Capacidad de pago calculada: ${capacidad_pago}")
        print(f"Puntuación de crédito: {puntuacion} puntos")
        print(f"Nivel de riesgo detectado: {nivel_riesgo}")
        print(f"DECISIÓN FINAL: {decision}")

    continuar = input("¿Desea realizar otra evaluación? (si/no): ").lower()

print("\nGracias por utilizar el Simulador de Buró de Crédito. Sesión finalizada.")
print("progama creado por nahum flores nl = 21")
```
